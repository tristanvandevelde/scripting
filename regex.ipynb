```python
number = "My number is 0486 96 41 83"
```

```python
digit = re.compile(r'\d')
digits = re.compile(r'\d+')
```

```python
digit.findall(number)
# [0,4,8,6,9,6,4,1,8,3]
```

```python
digits.findall(number)
# [0486,96,41,83]
```
