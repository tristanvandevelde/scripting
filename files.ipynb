## Compression

```python
import zipfile
```

```python
new_file = open('new_file.txt', 'w')
new_file.write("This is a compressed file")
new_file.close()
```

```python
surprise = open('surprise.py', 'w')
surprise.write("print('Hello, world!')")
surprise.close()
```

```python
os.mkdir("surprisedir")
```

```python
shutil.move("surprise.py", "surprisedir/surprise.py")
shutil.move("new_file.txt", "surprisedir/compressed.txt")
```

```python
zipfile_ = zipfile.ZipFile(Path("surprisedir/compressed.txt"))
```

Making a zip file:

```python
zipfile_ = zipfile.ZipFile("surprisedir.zip", "w")
zipfile_.write("surprisedir0/compressed.txt")
zipfile_.write("surprisedir0/surprise.py")
zipfile_.close()
```

Extracting a zip file:

```python
zipfile_ = zipfile.ZipFile(Path("surprisedir.zip"))
zipfile_.extractall()
zipfile_.close()
```
