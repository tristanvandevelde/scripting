## Exercise

Setting up the scraper:

```python
blockchain_url = "https://www.blockchain.com/btc/unconfirmed-transactions"
```

```python
blockchain_response = requests.get(blockchain_url)
```

```python
blockchain_html = blockchain_response.text
```

```python
blockchain_soup = BeautifulSoup(blockchain_html)
```

Scrape the hashes:

```python
blockchain_hash = blockchain_soup.find_all("a", {"class": "sc-1r996ns-0 fLwyDF sc-1tbyx6t-1 kCGMTY iklhnl-0 eEewhk d53qjk-0 ctEFcK"})
```

```python
hash_list = []
for h in blockchain_hash:
    hash_list.append(h.get_text())
```

Scrape the others (these have the same class):

```python
blockchain_other = blockchain_soup.find_all("span", {"class": "sc-1ryi78w-0 cILyoi sc-16b9dsl-1 ZwupP u3ufsr-0 eQTRKC"})
```

Scrape the time:

```python
blockchain_time = blockchain_other[::3]
```

```python
time_list = []
for t in blockchain_time:
    time = t.get_text()
    time_list.append(time)
```

Scrape the BTC value:

```python
blockchain_btc = blockchain_other[1::3]
```

```python
btc_list = []
for v in blockchain_btc:
    btc_list.append(v.get_text())
```

Scrape the USD value:

```python
blockchain_usd = blockchain_other[2::3]
```

```python
usd_list = []
for v in blockchain_usd:
    usd_list.append(v.get_text())
```

Combine into pandas dataframe:

```python
import pandas as pd
```

```python
bitcoin_df = pd.DataFrame(list(zip(blockchain_time, hash_list, blockchain_btc, blockchain_usd)),
                            columns=["Time", "Hash", "BTC", "USD"])
```
