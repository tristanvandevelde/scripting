## Multithreading

```python
import threading
```

```python
def Napping():
    time.sleep(5)
    print("Wake up!")
    
def Napping2():
    time.sleep(3)
    print("Wake up faster!")
```

```python
Napping()
print("start")
print("stop")
```

```python
print("Start!")
thread1 = threading.Thread(target=Napping)
thread2 = threading.Thread(target=Napping2)
thread1.start()
thread2.start()
print("Stop")
```
